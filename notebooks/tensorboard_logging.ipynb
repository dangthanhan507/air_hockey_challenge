```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_tensor_board_logs(log_path):
    ent_loss = []
    entropy = []
    l2_loss = []
    l2_norm = []
    loss = []
    neglogp = []
    
    with open(log_path, "r") as f:
        for i, line in enumerate(f.readlines()):
            divider_idx = line.find(" | ")
            space_idx = line.rfind(" ")
            if i % 14 == 4:
                ent_loss.append(float(line[divider_idx + 2:space_idx]))
            elif i % 14 == 5:
                entropy.append(float(line[divider_idx + 2:space_idx]))
            elif i % 14 == 7:
                l2_loss.append(float(line[divider_idx + 2:space_idx]))
            elif i % 14 == 8:
                l2_norm.append(float(line[divider_idx + 2:space_idx]))
            elif i % 14 == 9:
                loss.append(float(line[divider_idx + 2:space_idx]))
            elif i % 14 == 10:
                neglogp.append(float(line[divider_idx + 2:space_idx]))
            else:
                continue
    
    N = len(ent_loss)
    plt.figure(figsize=(10, 10))
    plt.subplot(321)
    plt.title("Entropy Loss")
    plt.plot(np.arange(N), np.array(ent_loss))
    plt.subplot(322)
    plt.title("Entropy")
    plt.plot(np.arange(N), np.array(entropy))
    plt.subplot(323)
    plt.title("L2 Loss")
    plt.plot(np.arange(N), np.array(l2_loss))
    plt.subplot(324)
    plt.title("L2 Norm")
    plt.plot(np.arange(N), np.array(l2_norm))
    plt.subplot(325)
    plt.title("Loss")
    plt.plot(np.arange(N), np.array(loss))
    plt.subplot(326)
    plt.title("Neg Log P")
    plt.plot(np.arange(N), np.array(neglogp))
```

```python
plot_tensor_board_logs("..b/train_output/run1")
```
